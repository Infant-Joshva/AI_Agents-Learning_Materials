```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
docs = ["AI is transforming the world", "Python is now powerful language because of the AI", "Claude is now ruling the IT compnies"]
docs_embedding = model.encode(docs)
```

```python
#setup

dimension = docs_embedding.shape[1]

#Need to create a object for index
index = faiss.IndexFlatL2(dimension)
index.add(np.array(docs_embedding))
```

```python
# Query calling
query = "What is AI"
query_embedding = model.encode(query)

#search
top_k = 2
distances, indices  = index.search(np.array(query_embedding).reshape(1,-1), top_k)
print(indices)
print(f"The top result is: ", [docs[i] for i in indices[0]])
```
