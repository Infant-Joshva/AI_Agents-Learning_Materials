```python
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings


# 1. Sample text input (could be chunks from PDFs, Notion, etc.)
docs = [
    Document(page_content="Python is a popular programming language for data science."),
    Document(page_content="LangChain is a framework to build applications using LLMs."),
    Document(page_content="Retrieval-Augmented Generation helps LLMs access external documents."),
    Document(page_content="FAISS is a vector store used for similarity search.")
]

# 2. Load Sentence Transformer model
model_name = "sentence-transformers/all-MiniLM-L6-v2"
embeddings = HuggingFaceEmbeddings(model_name=model_name)

# 3. Create FAISS vector store
vectorstore = FAISS.from_documents(docs, embeddings)

#4. convert to retriver
retrievers = vectorstore.as_retriever(search_type="similarity", k=2)

#5. Ask a query
query = "what helps LLMs retrieve external knowledge?"
result = retrievers.invoke(query)

#6. Show results
print(result)
for i, doc in enumerate(result):
  print(f"Relevant Doc {i+1}: \n{doc.page_content}\n{'-'*40}")
```
